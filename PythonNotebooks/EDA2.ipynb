```python
# define a function to create some data frames of just vehicle class counts aggregated by year
# I'm fairly certain there's an easier way to do this (by importing Counter for example)
# But I really wanted to solve this on my own
# the function filters by the 'model' passed to the function and then it groups by year
# it then groups by vehicle class and spits out value_counts()
# the last few steps are cleaning the value_counts() list and returning a good data frame
def model_time_plot(model):
    holder_1 = mpg[mpg['vehicle_class'] == model]
    holder_2 = holder_1.groupby('year')
    holder_3 = holder_2['vehicle_class'].value_counts()
    holder_4 = pd.DataFrame(holder_3)
    holder_4['year'] = holder_4.index
    holder_4.reset_index(drop=True, inplace=True)
    holder_4[['year', 'class']] = holder_4['year'].apply(pd.Series)
    
    return holder_4
```

```python
# function calls to create data frames for the vehicle classes I'm interested in looking at
plt1_df = model_time_plot('compact')
plt2_df = model_time_plot('subcompact')
plt3_df = model_time_plot('truck')
plt4_df = model_time_plot('suv')

# plot all the things
plt.plot('year', 'vehicle_class', data = plt1_df, marker = '.', 
         markerfacecolor = 'blue', markersize = 12, color = 'skyblue', linewidth = 2, label = 'compact')
plt.plot('year', 'vehicle_class', data = plt2_df, marker = '.', 
         markerfacecolor = 'darkgreen', markersize = 12, color = 'lightgreen', linewidth = 2, label = 'subcompact')
plt.plot('year','vehicle_class', data = plt3_df, marker = '.', 
         markerfacecolor = 'darkred', markersize = 12, color = 'red', linewidth = 2, label = 'truck')
plt.plot('year', 'vehicle_class', data = plt4_df, marker = '.', 
         markerfacecolor = '#e6e600', markersize = 12, color = '#ffff00', linewidth = 2, label = 'suv')

# plt.title('Vehicle Counts Over Time', fontsize = 24, fontweight = 'bold', color = 'lightgreen')
plt.xlabel('Year', color = 'white')
plt.ylabel('Vehicle Count', color = 'white')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors ='white')
plt.legend()

plt.savefig('../Presentation/image_files/class_time_clear.png', transparent = True, bbox_inches = 'tight')
plt.savefig('../Presentation/image_files/class_time.png', bbox_inches = 'tight')
plt.show();
```

# I lied - one more plot for use in reporting

A bubble plot the releationship between MPG, Displacement/Cylinders, and CO2.... Very illustrative.

The first plot below will be very illustrative for my report.



```python
# Create an grouped and aggregated data frame for plotting the relationship between MPG, Displacement, and CO2 emissions
h1_df = mpg_fin.groupby('displacement')
h2_df = h1_df[['combMPG', 'co2']].agg([np.mean])
h2_df.columns = ['MPG', 'CO2']
h2_df['displacement'] = h2_df.index

# Plot the results as a bubble chart
sns.lmplot(x="MPG", y="displacement", data=h2_df,
           fit_reg = False, scatter_kws={"s": h2_df.CO2*7, 'color': 'grey', 'alpha': '0.6'}, 
           size=7, aspect=1.5)

# plt.suptitle('The Relationship Between MPG, Displacement, and CO2 emissions', 
#              fontsize = 24, fontweight = 'bold',color = 'lightgreen')
plt.xlabel('MPG', color = 'white')
plt.ylabel('Displacement', color = 'white')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors ='white')

# output files for use in presentation
plt.savefig('../Presentation/image_files/smog1_clear.png', transparent = True, bbox_inches = 'tight')
plt.savefig('../Presentation/image_files/smog1.png', bbox_inches = 'tight')
```
