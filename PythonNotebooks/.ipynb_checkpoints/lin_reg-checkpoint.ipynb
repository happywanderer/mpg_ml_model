```python
# import all the things
import pandas  as pd 
import numpy  as np 
import pandas as pd
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import r2_score
from sklearn import model_selection
from sklearn.linear_model import Ridge

%matplotlib inline
```

```python
# read in polished data
mpg = pd.read_csv('../datasets/mpg_fin.csv')
```

```python
# a function that takes objects and turns them into categories
# the function also turns the categories into numeric values (assigning a simple 1up)
def cat_code(df):
    l_o_types = list(df.dtypes)
    col_names = list(df.columns)
    zip_list = list(zip(col_names, l_o_types))
    
    for i in range(len(zip_list)):
        if zip_list[i][1] == 'object':
            feat = zip_list[i][0]
            df[feat] = df[feat].astype('category')
        i += 1
    
    cat_columns = df.select_dtypes(['category']).columns
    cat_columns
    
    df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
```

```python
# pass the data frame to the cat_code function and peak at results
cat_code(mpg)
display(mpg.head().T)
```

```python
# do some statistical analysis using visual aids ... aka heatmap
# this is done as a function to cut down on typing 
def heat_map(df):
    
    # get correlation coefficients from data frame
    corr = df.corr()
    
    # plot all the things
    fig, ax = plt.subplots(figsize=(7, 7))
   
    #Generate Color Map, red & blue
    colormap = sns.diverging_palette(220, 10, as_cmap=True)
    
    #Generate Heat Map, allow annotations and place floats in map
    sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")
    
    #Apply xticks
    plt.xticks(range(len(corr.columns)), corr.columns);
    plt.yticks(range(len(corr.columns)), corr.columns)
    
    #show plot
    plt.show()
```

```python
# a function to scale data
# when using CO2 as a feature ... this becomes VERY important as CO2 values dwarf the rest
def scale_df(df):
      
    # use StandardScaler to morph the data    
    scaler = StandardScaler()
    scaler.fit(df)
    
    # assign the resulting data to new dataframe
    df_scaled = pd.DataFrame(data = scaler.transform(df), columns = df.columns, index = df.index)
    
    return df_scaled
```

```python
# A function that gets correlation coefficients and outputs in a user-friendly format
def get_coef(df):
    # create empty lists to house my values
    # and a list of columns
    coef_list = []   
    cols = []
    cols2 = list(df.columns)
    
    # loop through each column and append the correlation between that column and MPG into a list
    # I initially had a few hidden NaNs... so the try / except took care of that for me... 
    # with squeaky clean data, I wouldn't need try/except.
    for i in cols2:
        try:
            coef_list.append(abs(df.combMPG.corr(df[i])))
            cols.append(i)
        except TypeError:   
            pass
    
    coefs_df = pd.DataFrame(list(zip(cols, coef_list)), columns = ['feat', 'coef'])
#     coefs_df.rename(index = str, columns = {"0": "feat", "1": "coef"}, inplace = True)
    display(coefs_df.sort_values('coef', ascending=False))
    return coefs_df
```

```python
# pass the dataframe to the scale function and check on the output
# note - a critical mistake .. I scaled my target variable too.... 
# never do that
mpg_scaled = scale_df(mpg)
mpg_scaled.head(10)
```

```python
# pass the scaled data and non-scaled data to the get_coef function in order to .. get correlation coefficients
mpg_coef_df = get_coef(mpg)
mpg_sc_coef_df = get_coef(mpg_scaled)
```

```python
# a bit of code to plot the correlation coefficients in order to visual represent potential impact on predictions
# create a new data frame sans the combMPG ... cause we know it is 100% correlated to itself
h1_df = mpg_coef_df.drop(4).sort_values('coef', ascending = False)

# plot all the things
fig = plt.figure()
ax1 = fig.add_subplot(111)

h1_df.sort_values('coef').plot(kind = 'barh', x = 'feat', y = 'coef', 
                         color = 'skyblue', edgecolor = 'darkblue', ax = ax1, figsize = (10, 10))
# add some data labels
for i, v in enumerate(sorted(h1_df.coef)):
    plt.text(v+0.01, i, str(round(v, 2)), color='blue', va="center")

# add some text
plt.title('Impact of Variables on MPG', fontsize = 14, fontweight = 'bold')
plt.xlabel('Impact', fontsize = 14)
plt.ylabel('Variable', fontsize = 14)

# plt.savefig('../Presentation/image_files/top10_mpg_clear.png', transparent = True, bbox_inches = 'tight')
# plt.savefig('../Presentation/image_files/top10_mpg.png', bbox_inches = 'tight')
plt.savefig('../Presentation/image_files/coefs_clear', transparent = True, bbox_inches = 'tight')
plt.savefig('../Presentation/image_files/coefs', bbox_inches = 'tight')
plt.show()
```

```python
# Get baseline scores for my data
# Here I'm going to compare actual combMPG values to the mean value
# for combMPG (approx 20.8). This is like doing a linear regression test
# where all preditions are 20.8 ....
# if my actual linear regression models can't beat this baseline score...
# there are problems
def base_line(df):
    true = mpg_scaled.combMPG
    pred_base = np.full(true.shape[0], true.mean())
    
    print(metrics.mean_absolute_error(true, pred_base))
    print(metrics.mean_squared_error(true, pred_base))
    print(np.sqrt(metrics.mean_squared_error(true, pred_base)))
    print(r2_score(true, pred_base))
```

```python
# get baseline scores
base_line(mpg_scaled)
```

```python
# a function that does linear regression
def lin_reg(df, features):
    
    # this intermediate step isn't 100% necessary UNLESS I pass
    # all features from the full data frame (which would include years and combMPG)
    holder_df = df.drop(columns = ['combMPG'])
    
    # create a data frame based on features passed
    X = holder_df[features]
    y = mpg.combMPG   # remember my mistake of scaling my target variable... didn't use that here
    
    # split data into testing and training sets
    X_train, X_test, y_train, y_test = train_test_split(X, y)
    
    # instantiate lr
    lr = LinearRegression()
    model = lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)

  
    # generate a list of coefficients
    coef_df = (pd.DataFrame(list(zip(list(X.columns),lr.coef_))))
    coef_df.columns = ['feature', 'coefficient']
    coef_df['coefficient'] = coef_df.coefficient.abs()
            
    # Get the baseline score for comparison
    # Create a NumPy array with the same shape as y_test.
    y_null = np.zeros_like(y_test, dtype=float)

    # Fill the array with the mean value of y_test.
    y_null.fill(y_test.mean())
    
    # display the results (coefficients, intercept, MAE, MSE, RMSE, Null RMSE, and r^2 score)
    display(coef_df.sort_values('coefficient', ascending=False))
    print('\nThe intercept for this linear regression model: ', 
          lr.intercept_)
    print('The Mean Absolute Error (MAE) is: ', 
          metrics.mean_absolute_error(y_test, y_pred))
    print('The Mean Squared Error (MSE) is: ',
          metrics.mean_squared_error(y_test, y_pred))
    print('The Root Mean Squared Error (RMSE) is: ', 
          np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print('The baseline RMSE to beat is: ', 
          np.sqrt(metrics.mean_squared_error(y_test, y_null)))
    print('\nThe score for this linear regression model: ', 
          lr.score(X_test, y_test))
    
    # plot the predictions vs the actual
    plt.title('Linear Regression Plot')
    plt.scatter(y_test, y_pred)
    plt.xlabel('Measured')
    plt.ylabel('Predicted')
    plt.show()
    
    ###
    ## Cross Validation Section
    ###
    # kick off the cross validation by definig kf
    kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
    
    print("\nThe RMSE for the cross validation on the Training split is: ", 
          np.sqrt(np.mean(-cross_val_score(lr, X_train, y_train, cv=kf, scoring='neg_mean_squared_error'))))
    print("\nThe mean of R**2 on the Training split is: ", 
          np.mean(cross_val_score(lr, X_train, y_train, cv=kf)))
    
    print("\nThe RMSE for the cross validation on the Testing split is: ", 
          np.sqrt(np.mean(-cross_val_score(lr, X_test, y_test, cv=kf, scoring='neg_mean_squared_error'))))
    print("\nThe mean of R**2 on the Testing split is: ", 
          np.mean(cross_val_score(lr, X_test, y_test, cv=kf)))
    
    
    # Plot the cross validation predictions using the entire X_scaled dataframe and 
    predictions = cross_val_predict(lr, holder_df, y, cv=kf)
    plt.title('Cross Validation Plot')
    plt.scatter(y, predictions)
    plt.xlabel('True Values')
    plt.ylabel('Predictions')
    plt.show
```

```python
lin_reg(mpg_scaled, ['make', 'model', 'vehicle_class', 'cylinders',
               'displacement', 'drive', 'tran_type', 'fuel_type', 'co2', 'year'])
```

```python
# A function that does ridge regression + cross validation 
# in order to compare and contrast this with linear regression
def ridge_reg(df, features):
    # this intermediate step isn't 100% necessary UNLESS I pass
    # all features from the full data frame (which would include years and combMPG)
    holder_df = df.drop(columns = ['combMPG'])
    
    # create a data frame based on features passed
    X = holder_df[features]
    y = mpg.combMPG
    
    # split data into testing and training sets
    X_train, X_test, y_train, y_test = train_test_split(X, y)
    
    
    
    # Instantiate the model.
    # here I'm passing several alphas to see what will be the best
    ridgereg = Ridge(alpha=.0001)
        
    # Fit the model.
    model = ridgereg.fit(X_train, y_train)

    # Predict with fitted model.
    y_pred = ridgereg.predict(X_test)
    print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    
    # generate a list of coefficients
    coef_df = (pd.DataFrame(list(zip(list(X.columns),ridgereg.coef_))))
    coef_df.columns = ['feature', 'coefficient']
    coef_df['coefficient'] = coef_df.coefficient.abs()
    display(coef_df.sort_values('coefficient', ascending = False))
    
    # Print the resulting scores of the Ridge Regression Model
    print('The Mean Absolute Error (MAE) is: ', 
          metrics.mean_absolute_error(y_test, y_pred))
    print('The Mean Squared Error (MSE) is: ',
          metrics.mean_squared_error(y_test, y_pred))
    print('The Root Mean Squared Error (RMSE) is: ', 
          np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print('\nThe score for this linear regression model: ', 
          ridgereg.score(X_test, y_test))
    ###
    ## Cross Validation Section
    ###
    # kick off the cross validation by definig kf
    kf = model_selection.KFold(n_splits=15, shuffle=True)
    
    print("\nThe RMSE for the cross validation on the Training split is: ", 
          np.sqrt(np.mean(-cross_val_score(ridgereg, X_train, y_train, cv=kf, scoring='neg_mean_squared_error'))))
    print("\nThe mean of R**2 on the Training split is: ", 
          np.mean(cross_val_score(ridgereg, X_train, y_train, cv=kf)))
    
    print("\nThe RMSE for the cross validation on the Testing split is: ", 
          np.sqrt(np.mean(-cross_val_score(ridgereg, X_test, y_test, cv=kf, scoring='neg_mean_squared_error'))))
    print("\nThe mean of R**2 on the Testing split is: ", 
          np.mean(cross_val_score(ridgereg, X_test, y_test, cv=kf)))
    
    
    
    
```
