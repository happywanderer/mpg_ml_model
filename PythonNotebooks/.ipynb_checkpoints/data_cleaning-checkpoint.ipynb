## Deal with hybrid/electric vehicles present in the data

#### Here I made the command decision to simply dump these values versus analyze them with the data or separately 
I did this for several reasons:
1) potential for hybrids to skew data
2) time
3) future-proofing.

In 3, I mean that this model needs to be built for crunching data in the future when hybrids and evs will (we hope) make up a larger number of cars in data.  Therefore, their effect on skew will be greater. By separating now, we can prevent that.

```python
# do some quick digging because I noticed this data has hybrids/evs based on column names (and I read the data dictionary)
# create a dataframe that will help me identify the hybrids and evs
mpg_df2 = mpg[['phevBlended','tCharger', 'sCharger', 'atvType','evMotor', 'charge240b', 'phevCity', 'phevHwy','phevComb']]
# check the results
mpg_df2.atvType.value_counts()
```

```python
# createa new data frame that has NO hybrids or evs
mpg_df3 = mpg_df2[~mpg_df2.atvType.isin(['Hybrid', 'EV', 'Plug-in Hybrid'])]
# check the results
mpg_df3.phevBlended.value_counts()
```

```python
# and to keep naming simple - create a new mpg_df data frame with no hybrids/evs
mpg_df = mpg[~mpg.atvType.isin(['Hybrid', 'EV', 'Plug-in Hybrid'])]
mpg_df.fuelType.value_counts()
```

## Deal with bad columns identified in input step above

```python
# There were three bad columns identified on input of csv file
# this big of code identifies those columns and displays them
# in order to see why they were identified as bad and if they
# have any relevance on overall data
columns = list(mpg_df.columns.values)

col_list = [73, 74, 76]

# A pythonic way of assigning the actual column names to the bad_columns
bad_columns = [columns[x] for x in col_list]

print(bad_columns)   # Print the names of the bad columns

mpg_df_badcols = mpg_df[['mfrCode', 'c240Dscr', 'c240bDscr']]   # Make a dataframe out of the bad column data

# Display the data in the bad columns along with some preliminary EDA to determine if data is needed.
display(mpg_df_badcols.head())
print(mpg_df_badcols.shape)
print(mpg_df_badcols.isna().sum())   # This shows a ton of NaN values ... meaning these columns are basically worthless

# Drop the useless columns and never look back
mpg_df.drop(columns=bad_columns, inplace = True)
```

## Deal with null (NaN) values

A check of the data revealed that 3 columns had relatively few NaNs. Those rows were dropped.
One column, 'drive', had over 1000 NaNs.... EDA was conducted to determine a value to impute.

#### Note that evMotor columns has MANY NaN values - but I'll deal with that later by splitting electric vehicles into their own dataframe

```python
# Now the null values have to be checked for and dealt with
# There were some columns with very few null values. 
# Those could be dropped out without dramatic effect on data
# There was one column (drive) that had over 1000 null values.
# I felt dropping that much out would negatively impact data
# so I imputed a value for the NaNs with some preliminary EDA

print(mpg_df2.isnull().sum())

mpg_df2.dropna(subset=['cylinders', 'displacement', 'transmission'], inplace = True)   # Drop the rows with few NaNs

# EDA on 'drive' column to determine a good value to impute
print('\n', mpg_df2[mpg_df2['drive'].isnull() == True].year.value_counts())
print('\n',mpg_df2[mpg_df2.year == 1984].drive.value_counts())
```

```python
# I chose to use 2-Wheel Drive as it was the most common value
# Perhaps a more equitable solution would have been to assign
# a proportional number of 2-Wheel Drive and 4-Wheel of All-Wheel Drive
# values as they are relatively equal.

mpg_df2.drive.fillna(value='2-Wheel drive', inplace = True)

mpg_df2.isna().sum()    # Display results
```

## Clean up the drive values and combine a few that are ... in essence the same

I decided that 4-wheel drive and all-wheel drive are basically the same. The biggest difference being whether it is fulltime or parttime.... in the end... same thing.

```python
# Display the drive values for analysis and renaming
display(pd.DataFrame(mpg_df2.drive.value_counts()))

# rename 2-Wheel drive & 2-Wheel Drive to 2wd
# rename 4-Wheel or All-Wheel Drive to awd
# rename front-wheel drive to 2wd
# rename rear-wheel drive to 2wd
# rename part-time 4-wheel drive to awd
mpg_df2.drive.replace(['2-Wheel Drive', '2-Wheel drive'], '2wd', inplace = True)
mpg_df2.drive.replace(['4-Wheel or All-Wheel Drive', '4-Wheel Drive', 'Part-time 4-Wheel Drive', 'All-Wheel Drive'], 'awd', inplace = True)
mpg_df2.drive.replace(['Front-Wheel Drive'], 'fwd', inplace = True)
mpg_df2.drive.replace(['Rear-Wheel Drive'], 'rwd', inplace = True)

# Print the results of renaming
print(mpg_df2.drive.value_counts())
```

## Time to clean up the models list a bit

There is just way too much data to fully clean models, but below I've taken great strides to streamline many problems with data entries in the 'model' column. 

This bit actually took the longest to figure out. I had a partial solution that took WAY too long (an hour).

After a lot of trial and error, I found the best thing to do was what you see below.

```python
# This section of code is used to clean up the 'model' column a bit.
# This doesn't clean it completely as there is just too much junk.
# It does strip out some of the common junk in order to streamline models
# This will come into play a bit later in EDA and Machine Learning when
# 'model' is used as a slicer (along with make and year)

# Print the dirty list 
print('\nDirty model list:\n', mpg_df3.model.value_counts().head(10))

# split the model list on whitespace (once), 
# expand those values into individual columns, 
# and dump the whole thing into a new df then rename the new columns
df2 = mpg_df3.model.str.split(n = 1, expand = True)
df2.columns = ['model1', 'model2']

# create a new df that houses the original df plus the expanded 'model' columns
df3 = pd.concat([mpg_df3, df2], axis = 1)

# fill in NaN values 
# this is ridiculously important as NaNs cannot be iterated over
df3.model2.fillna(value = 'ZZZZZZ', inplace = True)

# 'model2' is still a list of long strings
# this step splits apart those strings into a list of lists
# output is stored in a list
model_list = list(df3.model2.str.split(expand = False))

# this list contains those 'junk' items in the 'model' column.
# again, this isn't all-inclusive
# this list was derived from continusouly printing value_counts() and identifiying new outliers
chars = ['2WD', '4WD', 'AWD', 'Pickup', '2x2', '4x4', '4x2', '(SOHC)', '(cargo)', 
         '(passenger)', 'FWD', 'RWD', '2Door', '4Door',
         '4/2','2.3CL/3.0CL', '2.5TL/3.2TL', '3.5RL', '3.0', 'A/C', 'Crew Cab', 
         'Super Cab', 'Boss', 'CNG', '16-Valve', '(Special Off-Road Model)', 'Bluetec',
         '(Bi-fuel)', '(Dedicated CNG)', '237HP', '4matic', 
         'Hatchback', 'TDV6', 'V8', 'MERC BENZ', 'V6', '4.0', '2.5', 
         '(94 Amp-hour battery)', 'Van', '286 Hp', 'SFE', 'XL/XLT', 'XFE', 'FFV',
         '5.0', '4.2', '2.8', 'Litre', 'eAssist', '4Matic', 'Cab Chassis', 'Supercharged', '5 Door',
         'xDrive', 'M Performance', 'Performance']

# And this section is where we rip the list of lists apart and replace the junk values
# The first for loop does the majority of the work by ripping the list of lists apart
# and then replacing items within the nested lists that match 'junk' characters
new_list = []
a = 0
b = 0
for a, item in enumerate(model_list):
    for b, word in enumerate(item):
        if word in chars:
            model_list[a][b] = model_list[a][b].replace(word, '')
        b += 1
    a += 1

# throw the result into a new list and then use a for loop to iterate through
# each item in the list of lists and join the individual peices into one
# string again
model_list2 = model_list
for a, item in enumerate(model_list2):
    model_list2[a] = ' '.join(item)    

# again creating a new list to conduct further operations
# this for loop strips whitespace at the beginning and end of each item in the list of strings
model_list3 = model_list2
for a, item in enumerate(model_list3):
    model_list3[a] = model_list3[a].strip()

# below I'm stripping any whitespace that might exist in 'model1' column
# then I'm overwriting 'model2' with the values in 'model_list3' list
df3['model1'] = df3.model1.str.strip()
df3['model2'] = model_list3

# Doing EDA on value_counts() revealed some whitespace rows
# this fills that whitespace so that the rows can be more easily
# worked with... 
df3.replace(r'^\s*$', 'ZZZZZZ', regex=True, inplace = True)

# overwriting 'model2' with model2 values after replacing the ZZZZZZ 
# inserted from the above step is a single whitespace 
df3['model2'] = df3.model2.str.replace('ZZZZZZ', ' ')

# creating a new 'model3' column in order to add the 'model1' and
# and 'model2' columns back together
df3['model3'] = df3.model1 + ' ' + df3.model2

# strip the whitespace again just in case
df3.model3 = df3.model3.str.strip()

# overwrite 'model' in the original df with our newly cleaned values
mpg_df3['model'] = df3.model3

# print the results
print('\nClean model list:\n', mpg_df3.model.value_counts().head(10))
```
